```python
headers ={'Token':'6112c32703f442f0'}
url = 'https://results.nyrr.org/api/runners/finishers-filter'
body = '{"eventCode":"M2019","runnerId":null,"searchString":null,"countryCode":null,"stateProvince":null,"city":null,"teamName":null,"teamCode":null,"gender":null,"ageFrom":null,"ageTo":null,"overallPlaceFrom":null,"overallPlaceTo":null,"paceFrom":null,"paceTo":null,"overallTimeFrom":null,"overallTimeTo":null,"gunTimeFrom":null,"gunTimeTo":null,"ageGradedTimeFrom":null,"ageGradedTimeTo":null,"ageGradedPlaceFrom":null,"ageGradedPlaceTo":null,"ageGradedPerformanceFrom":null,"ageGradedPerformanceTo":null,"handicap":null,"sortColumn":"overallTime","sortDescending":false,"pageIndex":1,"pageSize":50}'
```

- the `body` (copied from chrome devtools) is a string above, we need it as a dictionary so we can update the `pageIndex`
    - `json.loads()` converts the json string to json/dict
- the `url` (from chrome devtools is the url the API is hitting to get data)
- `headers` contain the API token (also harvested from devtools)

```python
body_dict = json.loads(body)
```

```python
body_dict
```

To make the http request, use the requests library's `post()` method - we can see it is a post by looking at devtools

```python
result = requests.post(url, json = body_dict, headers = headers)
result #Response 200 means ok
```

By drilling down into the json returned as result (`result.json()`) you can see that you need the `response` values and their `items`

```python
result_set = result.json()['response']['items']
#result.json() wil show first 50, have to nest down to response >items
```

```python
pd.DataFrame(result_set)
```

passing the dictionary (json) returned to `pd.DataFrame()` creates a df with the first 50 rows of runner data

```python
df = pd.DataFrame(result_set)
```

```python
df.tail()
```

```python
#for loop to get each set of results per page of url
df = pd.DataFrame()
for index in range (1, 1072): #or can do (pages+1 as end number)
    #print(index)
    body_dict['pageIndex']=index
    result = requests.post(url, json = body_dict, headers = headers)
    result_set = result.json()['response']['items']
    df = df.append(pd.DataFrame(result_set))
    print(df.shape) #or run tail here
```

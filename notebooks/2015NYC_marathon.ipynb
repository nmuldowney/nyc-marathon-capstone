```python
#for loop to get each set of results per page of url
df = pd.DataFrame()
for index in range (1, 971): #or can do (pages+1 as end number)
    #print(index)
    body_dict['pageIndex']=index
    result = requests.post(url, json = body_dict, headers = headers)
    result_set = result.json()['response']['items']
    df = df.append(pd.DataFrame(result_set))
    print(df.shape) 
```
